```python
from chembl_webresource_client.new_client import new_client
```

```python
#target searching
tar = new_client.target
query = tar.search('coronavirus')
targets = pd.DataFrame.from_dict(query)
targets
```

# select and retrieve bioactivity data

```python
selected_data = targets.target_chembl_id[1]
selected_data
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_data).filter(standard_type="IC50")
res
```

```python
da = pd.DataFrame.from_dict(res)
da
```
